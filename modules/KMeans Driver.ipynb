# This Notebook is a demo of my KMeans implementation.

```python
from kmeans import kmeans
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import random
import numpy as np
import sklearn

%matplotlib inline
```

### Let's make some clusters with make_blob

```python
    X, assignment= make_blobs(n_samples=[213, 124, 122, 194], cluster_std=0.80, random_state=7)

    
```

### Here's a scatterplot of the clusters colored by their true cluster assignment (from make_blobs)

```python
plt.scatter(X[:,0], X[:,1], c=assignment, cmap='gist_rainbow')
```

### My version of KMeans - UNBALANCED

```python
#Creating an instance of KMeans with 4 clusters and the default of 100 max iterations
km = kmeans(k=4, balanced=False)

#Let's get the output of fitting kmeans to X
output = km.fit(X)

#First thing in the array is a list of cluster assignments
#nth data point in X is assigned to the cluster in the nth index of this list
cluster_assignments = output[0]

#Second thing in the array is a list of the centers of the clusters
cluster_centroids = np.array(output[1])
```

```python
#Let's draw all the datapoints in X colored by their cluster assignments
plt.scatter(output[2][:,0], output[2][:,1], c=cluster_assignments, cmap='jet')

#Let's draw a red cross for the center of each cluster
plt.scatter(cluster_centroids[:, 0], cluster_centroids[:, 1], s=500, c='red', marker="x")
plt.show()
```

### If we observe the scatterplot above, we clearly see that the centroids have converged correctly, if they haven't converged correctly at the first try, they will on the second or third try.

```python
for x in np.unique(cluster_assignments):
    print("Items in cluster {}: {}".format(x, cluster_assignments.count(x)))
```

### My version of KMeans - BALANCED


```python
#Creating an instance of KMeans with 4 clusters and the default of 100 max iterations
km_balanced = kmeans(k=4, balanced=True)

#Let's get the output of fitting kmeans to X
output_balanced = km_balanced.fit(X)

#First thing in the array is a list of cluster assignments
#nth data point in X is assigned to the cluster in the nth index of this list
cluster_assignments_balanced = output_balanced[0]

#Second thing in the array is a list of the centers of the clusters
cluster_centroids_balanced = np.array(output_balanced[1])
```

```python
#Let's draw all the datapoints in X colored by their cluster assignments
plt.scatter(output_balanced[2][:,0], output_balanced[2][:,1], c=cluster_assignments_balanced, cmap='jet')

#Let's draw a red cross for the center of each cluster
plt.scatter(cluster_centroids_balanced[:, 0], cluster_centroids_balanced[:, 1], s=500, c='red', marker="x")
plt.show()
```

### If we observe the scatterplot above, we clearly see that the centroids have converged correctly, if they haven't converged correctly at the first try, they will on the second or third try. Furthermore, clusters are balanced, meaning that they contain roughly the same amount of datapoints (precise to 1 datapoint) .

```python
for x in np.unique(cluster_assignments_balanced):
    print("Items in cluster {}: {}".format(x, cluster_assignments_balanced.count(x)))
    
```
